```python
from astropy.io import fits
```

```python
from astropy.nddata import Cutout2D
from astropy.convolution import convolve, Box2DKernel, convolve_fft
from scipy.signal import convolve as sciconvolve
import matplotlib.pyplot as plt

```

```python
rband = fits.open('/content/drive/MyDrive/DESJ053816.9-503050.8_r.fits')

data = rband[0].data

axis = plt.subplot(111)
center = (550, 550)
size = (200, 200)
imr = axis.imshow(data, origin='lower', interpolation = 'nearest')
cutout = Cutout2D(data, center, size)
plt.imshow(cutout.data, origin = 'lower')
plt.title('Original Image 200x200')
plt.xlabel('x pixels')
plt.ylabel('y pixels')
```

```python
kernel = Box2DKernel(3)
astro3 = convolve(data, kernel)

axis1 = plt.subplot(111)
image1 = axis1.imshow(astro3, origin = 'lower', interpolation = 'nearest')
cutout33 = Cutout2D(astro3, center, size)
plt.imshow(cutout33.data, origin = 'lower')
plt.title('3x3 Centered at 200x200')
plt.xlabel('x pixels')
plt.ylabel('y pixels')
```

```python
kernel77 = Box2DKernel(7)
astro7 = convolve(data, kernel77)

axis2 = plt.subplot(111)
image2 = axis2.imshow(astro7, origin = 'lower', interpolation = 'nearest')
cutout77 = Cutout2D(astro7, center, size)
plt.imshow(cutout77.data, origin = 'lower')
plt.title('7x7 Centered at 200x200')
plt.xlabel('x pixels')
plt.ylabel('y pixels')
```

```python
kernel11 = Box2DKernel(11)
astro11 = convolve(data, kernel11)

axis11 = plt.subplot(111)
image11 = axis11.imshow(astro11, origin = 'lower', interpolation = 'nearest')
cutout11 = Cutout2D(astro11, center, size)
plt.imshow(cutout11.data, origin = 'lower')
plt.title('11x11 Centered at 200x200')
plt.xlabel('x pixels')
plt.ylabel('y pixels')
```
