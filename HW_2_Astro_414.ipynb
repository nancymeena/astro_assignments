```
#1
import numpy as np
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
from matplotlib.patches import Polygon


ntot, binsize = 1000000, 0.5
mean = 10
std= 2
x = mean + std*np.random.randn(ntot)  #100 samples with a mean of 10 and stdv of 25
bins = np.arange(0, 20, 0.5)

plt.figure(figsize=(10, 6))

# the histogram of the data
y, bins, patches = plt.hist(x, bins = bins, facecolor = 'r')
ybin = np.size(y)

bincenter = (bins + 0.25)[0:ybin]
er = np.sqrt(y) # gives error for the plot, square root of the data

gaus = ntot/(std*np.sqrt(2 * np.pi)) * binsize
plt.plot(bincenter, gaus * np.exp(-np.power(bincenter - mean, 2.)/2*np.power(std, 2)), 'b')

#labels
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.title('Gaussian Distribution')

#error bars
plt.errorbar(bincenter, y, yerr = er, marker ='x')
```
