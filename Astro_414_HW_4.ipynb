```
data =ascii.read('/content/gdrive/MyDrive/mktrans_zm_10_10.dat.txt')
os.path.isdir('/content/gdrive/MyDrive/filters')

path = "/content/gdrive/MyDrive/filters"

# all the files in filters folder.
filter1 =ascii.read('/content/gdrive/MyDrive/filters/filters/f098m.IR.tab.txt')
filter2= ascii.read('/content/gdrive/MyDrive/filters/filters/f105w.IR.tab.txt')
filter3= ascii.read('/content/gdrive/MyDrive/filters/filters/f110w.IR.tab.txt')
filter4= ascii.read('/content/gdrive/MyDrive/filters/filters/f125w.IR.tab.txt')
filter5= ascii.read('/content/gdrive/MyDrive/filters/filters/f140w.IR.tab.txt')
filter6= ascii.read('/content/gdrive/MyDrive/filters/filters/f160w.IR.tab.txt')

```

```
wave=data['col1']
trans= data['col2']

plt.plot(wave, trans, 'c')
plt.xlim(0.9, 2.7)
plt.ylim(0, 1.05)

plt.xlabel(r"wavelength[$\mu$m]")
plt.ylabel(r'transmission')
plt.title('Atmospheric Transmission Curve Near-IR vs HST-IR Filters')

xfilt1 = filter1['col2']/10000
yfilt1 = filter1['col3']
plt.plot(xfilt1, yfilt1, 'b', label = 'f098m')

xfilt2 = filter2['col2']/10000
yfilt2 = filter2['col3']
plt.plot(xfilt2, yfilt2, 'orange')

xfilt3 = filter3['col2']/10000
yfilt3 = filter3['col3']
plt.plot(xfilt3, yfilt3, 'g')

xfilt4 = filter4['col2']/10000
yfilt4 = filter4['col3']
plt.plot(xfilt4, yfilt4, 'r')

xfilt5 = filter5['col2']/10000
yfilt5 = filter5['col3']
plt.plot(xfilt5, yfilt5, 'purple')

xfilt6 = filter6['col2']/10000
yfilt6 = filter6['col3']
plt.plot(xfilt6, yfilt6, 'maroon')
```
